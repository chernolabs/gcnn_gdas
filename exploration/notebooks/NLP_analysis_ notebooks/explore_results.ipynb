```python
ngram_dict = {0:"monogramas",1:"bigramas",2:"trigramas",3:"mixto"}

def supergrafico(df,i,titulo):
    entropy_col = f"entropia_{i}"
    meansim_col = f"mean_sim_lsa_{i}"
    x = 1-df[entropy_col] 
    fig = px.scatter(df, x=x, y=meansim_col, size="tama単o",width=1000, height=600, title=f"Especificidad vs similaridad media vs tama単o - {titulo}",marginal_x="histogram",marginal_y="histogram",range_x=[0,1],range_y=[0,1])
    fig.update_layout(xaxis_title="Spec", yaxis_title=f"Similaridad media - {ngram_dict[i]}")

    fig.show()
```

```python
supergrafico(infomap_summary,0,"Infomap")
```

```python
supergrafico(louvain_summary,0,"Louvain")
```

```python
import pandas as pd
import json
data_folder = "../../../data/external/"
reports_folder = "../../../reports/model_predictions/"
```

```python
data = json.load(open(data_folder+'genomic_neighbors.json'))
```

```python
num_values_df = pd.DataFrame.from_dict({k:len(v) for k,v in data.items()}, orient="index",columns=["num_values"])

#había algunos que no tienen 100 y eso me rompe el dataframe asi que no los conte
dif_value = num_values_df[num_values_df.num_values != 100].index.values
num_values_df = num_values_df.drop(dif_value,axis=0)

df_data = []
for k in num_values_df.index.values:
    df_data.append(data[k])

genomic_n = pd.DataFrame(df_data, index = num_values_df.index.values)
genomic_n
```
